````python
import numpy as np

with open('rawData/del_ex_11_CP_OMO.data', 'r') as fStream:
    dati = fStream.read()
    fStream.close()
````

````python
# print (dati)

startStr = dati.find("'DATA9': (") + len("'DATA9': (")
stringaData9 = dati[startStr:]
stopStr = stringaData9.find(')')
# print  (stringaData9[:stopStr])
stringaData9 = stringaData9[:stopStr]
elenco = stringaData9.split(',')
print("Length elenco: {}".format(len(elenco)))
````

````python
DATA9 = []
for el in elenco:
    DATA9.append(float(el.strip()))
DATAG = np.array(DATA9, dtype=float)
print("Guanine vector: {}\nLength vector: {}\nDims: {}".format(DATAG, len(DATAG), DATAG.shape))
````

Attempts to find the peaks in a 1-D array.

The general approach is to smooth vector by convolving it with wavelet(width) for each width in widths. Relative maxima which appear at enough length scales, and with sufficiently high SNR, are accepted.

````python
for query in list_of_queries:
    print("Request for {}\n".format(query))
    handle = Entrez.esearch(db="nucleotide", retmax=10, term="{} AND COI".format(query))
    record = Entrez.read(handle)
    handle.close()
    try:
        id_ = record['IdList'][0]
        data = Entrez.efetch(db="nucleotide", id = id_, retmode='xml')
        extracted = Entrez.read(data)
        print(">{} {}\n{}\n\n".format(extracted[0]['GBSeq_accession-version'],
                              extracted[0]['GBSeq_definition'],
                             str(extracted[0]['GBSeq_sequence']).upper()))
        with open("fastaFiles/{}.fst".format(query), 'w') as file:
            file.write(">{} {}\n{}\n\n".format(extracted[0]['GBSeq_accession-version'],
                              extracted[0]['GBSeq_definition'],
                             str(extracted[0]['GBSeq_sequence']).upper()))
        file.close()
    except:
        print("No records found for {}\n\n".format(query))
        pass
    time.sleep(5)
````
